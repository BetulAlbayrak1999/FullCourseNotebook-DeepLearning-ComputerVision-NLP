```python
ImageArray=[]
W, H= 100, 100
for filename in os.listdir(ImagePath):
  img_path= os.path.join(ImagePath, filename)
  img= cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
  if img is not None:
    img= cv2.resize(img, (H, W))
    img= img.reshape(H, W, 1)
    ImageArray.append(img)

len(ImageArray)
```

```python
y= [np.random.choice([0, 1]) for i in range(250)]
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(np.array(ImageArray), np.array(y), test_size= 0.25, random_state=44, shuffle=True)

print('X_train shape is: ', X_train.shape)
print('y_train shape is: ', y_train.shape)
print('X_test shape is: ', X_test.shape)
print('y_test shape is: ', y_test.shape)
```

```python
KerasModel= keras.models.Sequential([
    keras.layers.Conv2D(filters=32, kernel_size=(3,3), strides= (1, 1), padding='VALID', input_shape=(H, W, 1)), #SAME
    keras.layers.MaxPooling2D(pool_size=(2, 2), strides=(2, 2)),
    keras.layers.BatchNormalization(),
    keras.layers.Conv2D(filters=32, kernel_size=(3, 3), strides=(1, 1), padding='VALID'),
    keras.layers.MaxPooling2D(pool_size= (2, 2), strides=(2, 2)),
    keras.layers.BatchNormalization(),
    keras.layers.Flatten(),
    keras.layers.Dense(units=128, activation=tf.nn.relu),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(1, activation=tf.nn.sigmoid)
])
```

```python
KerasModel.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
KerasModel.fit(X_train, y_train, epochs=10) #, batch_size=32 ,validation_data=(X_test, y_test)
```
